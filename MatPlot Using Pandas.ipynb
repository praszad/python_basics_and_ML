```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]","").str[:-2].astype(int)
car_sales
```

```python
car_sales["Total Price"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Make",y="Odometer (KM)",kind="bar");
```

```python
cs = pd.DataFrame(car_sales)
cs.plot.scatter(x="Odometer (KM)",y="Price")
```

```python
car_sales.plot(x="Odometer (KM)",kind="hist")
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease["age"].hist()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# over_50.plot.hist(figsize=(10,30))
fig , ax = plt.subplots()
ax.hist(over_50);
```

```python
over_50.plot(x="age",y="thalach",c="target",kind="scatter",figsize=(10,6))
```

```python
fig , ax = plt.subplots()
over_50.plot(x="age",y="chol",c="target",kind="scatter",ax=ax,figsize=(10,6))
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
ax.set(title="Heart Disease Scatter Plot",xlabel="Age",ylabel="Choloesterol")
ax.axvline(over_50["age"].mean())
ax.legend(*scatter.legend_elements(),title="target")
```

```python
fig , (ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
ax0.set(title="Heart Disease Plot 1",xlabel="Age",ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(over_50["chol"].mean())
scatter1 = ax1.scatter(x=over_50["age"],y=over_50['thalach'],c=over_50["target"])
```
